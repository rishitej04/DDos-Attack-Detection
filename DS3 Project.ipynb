```python
# Dropping the Object Data and Nan Value rows:

df = df.drop(['src'], axis = 1)
df = df.drop(['dst'], axis = 1)
df = df.drop(['Protocol'], axis = 1)
df = df.dropna()

columns_ini = list(df.columns)
x = df.drop(['label'], axis = 1)
y = df['label']

#Normalising the Data 

columns = list(x.columns)
scaled_data = preprocessing.normalize(x)
scaled_data = pd.DataFrame(x, columns = columns)
scaled_data.head()
```

```python
#Splitting the Data into Train and Test Sets

warnings.filterwarnings("ignore",)
X_train,X_test,Y_train,Y_test = train_test_split(scaled_data, y, test_size = 0.3, random_state = 0)
```

```python
# Plotting Heat Map of Covaraince 

plt.figure(figsize=(12,7))
cor = X_train.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.inferno)
plt.show()
```

```python
# with the following function we can select highly correlated features
# it we remove the first feature that is correlated with anything other feature

def correlation(dataset, threshold):
    col_corr = set() # set of all the names of correlated columns    
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs (corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i] # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
# Setting Correlation Threshold Value to Drop the columns

correlation(X_train, 0.7)
```

```python
corr_features = correlation(X_train, 0.7)

# Dropping the Columns which match the Threshold Value

x_new_train = X_train.drop(corr_features, axis=1)
x_new_test = X_test.drop(corr_features, axis=1)
```

```python
#KNN Method

def KNN():
    neighbours = 10
    test_acc = []
    
    for i in range(1,neighbours):
        knn = KNC(n_neighbors = i) #Describing the KNN model
        knn.fit(x_new_train,Y_train)      
        test_acc.append(knn.score(x_new_test, Y_test)) #Testing the model  
        
    plt.figure(figsize = (12,7), dpi = 80)#Plotting the accuracy for each value of neighbours
    plt.plot(range(1,neighbours), test_acc, label = 'Testing dataset Accuracy')   
    plt.legend()
    plt.xlabel('n_neighbors')
    plt.ylabel('Accuracy')
    plt.show()

    return('Maximum accuracy of KNN is ' +  str(max(test_acc)*100) + '%')
```

```python
# K-Means Method

def K_means():
    kmeans = KMeans(n_clusters = 2)
    kmeans.fit(x_new_train)
    Y_pred = kmeans.predict(x_new_test)
    centers = kmeans.cluster_centers_
    
    return('Accuracy of the K_means is ' + str(accuracy_score(Y_test, Y_pred)*100) + "%")
```

```python
# Naive Bayes Method

def NaiveBayes():
    gnb = GaussianNB()    
    gnb.fit(x_new_train, Y_train)    
    y_pred = gnb.predict(x_new_test)
    
    return('Accuracy of the Naive Bayes model is ' + str(accuracy_score(Y_test, y_pred)*100) + "%")
```

```python
# Support Logistic Regression Method

def LogReg():
    LR = LogisticRegression(random_state = 0, solver = 'newton-cg', max_iter = 5000)
    LR.fit(x_new_train, Y_train)    
    Y_pred = LR.predict(x_new_test)
    
    return('Accuracy of the Logistic Regression is ' + str(accuracy_score(Y_test, Y_pred)*100) + "%")
```

```python
# Hierachical Clustering Method 

def Hierarchichal():
    hierarchical_cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
    Y_pred = hierarchical_cluster.fit_predict(x_new_test)
    
    return('Accuracy of the Hirarchichal is ' + str(accuracy_score(Y_test, Y_pred)*100) + "%")
```
